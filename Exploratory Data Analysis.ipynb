## Feature Clustering:
    Let's cluster the latitutde and longitude feature of the dataset.

```python
X = drop_rows_data[['lat','lng']]
```

```python
X['lat'] = (X.lat-X.lat.mean())/X.lat.std()
```

```python
X['lng'] = (X.lng-X.lng.mean())/X.lng.std()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, max_iter=20)
X['cluster'] = kmeans.fit_predict(X)
X['cluster'] = X['cluster'].astype('category')
```

```python

sns.relplot(x = 'lng', y = 'lat', hue = 'cluster', data = X, height=6)
```

```python
X['Status'] = drop_rows_data['status']
```

```python
sns.catplot(x = 'Status', y = 'cluster', data = X, height=6)
```
